```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('./csv/train.csv',encoding='utf-8')
df_test = pd.read_csv('./csv/test.csv',encoding='utf-8')
df_weather = pd.read_csv('./csv/weather.csv',encoding='utf-8')
```

```python
print(df_train.shape)
print(df_test.shape)
```

```python
df_weather['year'] = df_weather['date'].apply(lambda x: int(str(x)[:4]))
df_weather['month'] = df_weather['date'].apply(lambda x: int(str(x)[4:6]))
```

```python
agg_cols = ['mean_temp','max_temp','min_temp','sum_rain','sun_time','mean_humid']
gb_df = df_weather.groupby(['area','year','month'])[agg_cols].agg(['mean','max','min']).reset_index()

new_cols = []
for col1,col2 in gb_df.columns:
    if col2:
        new_cols.append(col2+'_'+col1)
    else:
        new_cols.append(col1)
gb_df.columns = new_cols
gb_df.head()
```

```python
agg_cols
```

```python
agg_cols = [i for i in gb_df.columns if i not in['year','month','area']]
tmp_df = gb_df.groupby(['year','month'])[agg_cols].agg(['mean']).reset_index()
tmp_df.head()
new_cols = []
for col1,col2 in tmp_df.columns:
    new_cols.append(col1)

tmp_df.columns = new_cols
tmp_df['area'] = '全国'
tmp_df = tmp_df[gb_df.columns]
tmp_df.head()
```

```python
df_weather = pd.concat([gb_df,tmp_df])
df_weather.head()
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
all_df = pd.concat([df_train,df_test])
```

```python
area_pairs = all_df['area'].unique()
yasai_areas = set()

for area_pair in area_pairs:
    areas = area_pair.split('_')
    yasai_areas |= set(areas)

yasai_areas
```

```python
we_areas = df_weather['area'].unique()
we_areas
```

```python
area_map = {}

update_area_map = {'岩手':'盛岡','宮城':'仙台','茨城':'水戸','栃木':'宇都宮','群馬':'前橋','埼玉':'熊谷','山梨':'甲府',
                   '神奈川':'横浜','静岡':'浜松','愛知':'名古屋','香川':'高松','愛媛':'松山','沖縄':'那覇','兵庫':'神戸',
                   '北海道':'帯広',
                  }

for yasai_area in yasai_areas:
    if yasai_area not in we_areas and yasai_area not in update_area_map:
        area_map[yasai_area] = '全国'
    else:
        area_map[yasai_area] = yasai_area

area_map = {**area_map,**update_area_map}
area_map
```

```python
all_df
```

```python
all_df['area'] = all_df['area'].apply(lambda x: '_'.join([area_map[i] for i in x.split('_')]))
all_df
```

```python
all_df.shape
```

```python
df_train.shape[0]
```

```python
test_df = all_df.iloc[df_train.shape[0]:]
train_df = all_df.iloc[:df_train.shape[0]]
```

```python
test_df.to_csv('preprocessed_test.csv', index=False)
train_df.to_csv('preprocessed_train.csv', index=False)
```

```python
area_pairs = all_df['area'].unique()
```

```python
target_cols = [i for i in df_weather.columns if i != 'area']
```

```python
area_pair_dfs = []

for area_pair in area_pairs:
    areas = area_pair.split('_')
    if len(areas) > 0:
        area = areas[0]
        base_tmp_df = df_weather[df_weather['area'] == area]
        base_tmp_df = base_tmp_df[target_cols].reset_index(drop=True)
        for area in areas[1:]:
            tmp_df = df_weather[df_weather['area'] == area]
            tmp_df = tmp_df[target_cols].reset_index(drop=True)
            base_tmp_df = base_tmp_df.add(tmp_df)
        base_tmp_df /= len(areas)
        base_tmp_df['area'] = area_pair
        area_pair_dfs.append(base_tmp_df)
```

```python
area_pair_df = pd.concat(area_pair_dfs)
print(area_pair_df.shape)
area_pair_df.head()
```

```python
area_pair_df.to_csv('./preprocessed_weather.csv', index=False)
```
